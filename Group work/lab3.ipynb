```python
# change these parameters to more realistic values
mu = 0.3
L = 3
g = 9.8
```

```python
# To make use of scipy.optimize curve_fit this is the form we should present the model
# watch https://www.youtube.com/watch?v=tl5QNhSe0Yk to fresh up your knowledge on curve fitting
def model(X, mu, L, g):
    ''' return the angular acceleration of the pendulum theta is the angle, thata dot the angular velocity''' 
    theta, theta_dot = X
    return -mu * theta_dot - (g/L) * np.sin(theta) # = theta double dot 
```

```python
# Solving the angular behaviour of the pendulum numerically for an given input time t (in seconds).
# From now we will call this the simulation

def get_theta_double_dot_NN(x_y_z,value_to_predict):
    # unfold x_y_z
    x,y,z= x_y_z
    X=list(zip(x,y))
    Y=z
    clf = MLPRegressor(solver='lbfgs', alpha=1e-7,max_iter=500, hidden_layer_sizes=(50,20), random_state=1)
    clf.fit(X,Y)
#     predict_value = clf.predict(X)
    return clf.predict([value_to_predict])

def state(t,x_init,y_init,x_y_z,value_to_predict):  
    delta_t =0.01 #choose a delta in time to compute
    theta=x_init
    theta_dot=y_init
    for time in np.arange(0,t,delta_t):
        theta_double_dot = get_theta_double_dot_NN(x_y_z,value_to_predict)
        theta += theta_dot * delta_t
        theta_dot += theta_double_dot * delta_t
    return theta, theta_dot

def get_delta(x,y,z,X,U,V):
    #allocate in the memory for theta, theta_dot, delta_theta, delta_theta_dot
    x_y_z=[x,y,z]
    theta= np.zeros(U.shape[0])
    theta_dot= np.zeros(V.shape[0])
    delta_theta= np.zeros(U.shape)
    delta_theta_dot= np.zeros(V.shape)
    for i in range (U.shape[0]):
        for j in range (V.shape[0]):
            theta[i],theta_dot[j]=state(t,U[i][j],V[i][j],x_y_z,X[j])
            delta_theta[i][j]= theta[i]-U[i][j]
            delta_theta_dot[i][j]= theta_dot[j]-V[i][j]
    return delta_theta,delta_theta_dot

### initial an array for THETA_0, THETA_DOT_0

# THETA_0 =np.linspace(-2*np.pi,2*np.pi,30)
# THETA_DOT_0= np.linspace(-10, 9,30)
t=0.001

U, V = np.meshgrid(x,y)
print(U.shape)

print(V.shape)


delta_theta,delta_theta_dot=get_delta(x,y,z,X,U,V)
        
print(delta_theta.shape)      
print(delta_theta_dot.shape)      
```

```python
def get_theta_double_dot_with_a(theta, theta_dot, a_motor):
    # angular acceleration of the pendulum
    return -mu*theta_dot - ((g/L)*np.sin(theta)) - ((a_motor/L)*np.cos(theta))
```

```python
class Motor_Acc:
    def __init__(self, direction, speed, duration):
        self.direction = direction
        self.speed = speed
        self.duration = duration
        self.time = 0
        self.func = self.create_func()
    
    def create_func(self):
        ''' Create a list of acceleration values so that every step correspons to a step in simulation.
        Use the direction, speed and duration to create this list.
        In a next Lab we will use the data of the prototype to fit a function. 
        If you want to create a function now, that is fine.
        '''
        func_list=list()
        delta_t= self.duration/100
        for i in np.arange(0,self.duration,delta_t):
            motor_acc = self.direction*(self.speed/delta_t)   # motor acceleration
            func_list.append(motor_acc)
        return func_list #or some function
        
    def tik(self):
        ''' return the acceleration at a given time and shift the time '''
        acceleration = 0        
        if self.time < self.duration:
            acceleration = self.func[self.time]
        self.time += 1
        return acceleration
    
    def set_motor(self, direction, speed, duration):
        self.direction = direction
        self.speed = speed
        self.duration = duration
        self.time = 0
        self.func = self.create_func()
```

This video might help you to understand PID: https://www.youtube.com/watch?v=UR0hOmjaHp0  
Now the problem of control is a continues one so we can ignore using the motor class for now.

```python
def pid_control(t, THETA_0,THETA_DOT_0, a, goal):
    theta = THETA_0
    theta_dot = THETA_DOT_0
    list_t = []
    list_e = []
    error_prev = 0
    error_sum=0
    
    delta_t = 0.01#choose a delta in time to compute
    for time in np.arange(0,t,delta_t):
        theta_double_dot = get_theta_double_dot_with_a(theta, theta_dot, a)
        theta += theta_dot * delta_t
        theta_dot += theta_double_dot * delta_t
        
        error = goal-theta
        error_sum += error
        
        p = 9.9
        i = 3
        d = 9.9
        
        if len(list_t)>0:
            # We need at least one step in simulation to be able to make use of an Integral or derivative component.
            a = p*error + i*(error_sum) + d*(error-error_prev)
        else:
            # the first step can only be a Proportional change
            a = p*error
        list_t.append(theta)
        list_e.append(error)
        error_prev = error
        #print(a)
    return list_t, list_e

#lest say we start with an acceleration of 0
# for now we manipulate a directly to get an idea of how we should control things.
a = 0
#simulation_time = 60
simulation_time = 5
# we now simulate what happens over time with theta and the error. 
# The error is the difference between the goal of pi radius and the actual theta. 
theta_list,error_list = pid_control(simulation_time, np.pi-0.03, -0.01, a, np.pi)

plt.plot(range(len(theta_list)), theta_list)
plt.show()
```
